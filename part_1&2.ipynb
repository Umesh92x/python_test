## Moving average

```python
weeks = [4, 16, 28, 40, 52]
```

```python
dummy_var = pd.DataFrame()
#resample into weeks
dummy_var['Close'] = tcs['Close'].resample('W').mean() 
for i in range(len(weeks)):
    MA = dummy_var['Close'].rolling(weeks[i]).mean()
    dummy_var['Mov_avg'+str(weeks[i])] =MA
    print(dummy_var['Close'])
dummy_var.plot(title="Moving Averages for {} \n\n" .format('tcs'))
```

### lets do it through a function...

```python
def moving_avg(stock_name , weeks=[4,16,32,40,52]):
    dummy_var = pd.DataFrame()
#resample into weeks
    dummy_var['Close'] = stock_name['Close'].resample('W').mean() 
    for i in range(len(weeks)):
        MA = dummy_var['Close'].rolling(weeks[i]).mean()
        dummy_var['Mov_avg'+str(weeks[i])] =MA
        print(dummy_var['Close'])
    dummy_var.plot(title="Moving Averages for {} \n\n" .format('infy'))
moving_avg(stock_name)
```

# Volume shoke

```python
def volume_shock(stock_name):
    stock_name["vol_T+1"] = stock_name.Volume.shift(1)
    stock_name["volume_shock"] = ((abs(stock_name["vol_T+1"] - stock_name["Volume"])/stock_name["Volume"]*100)  > 10).astype(int)
    return stock_name
```

```python
volume_shock(tcs)
```

```python
volume_shock(infy)
```

```python
volume_shock(nifty)
```

```python
def price_shock(stock_name):
    stock_name["price_T+1"] = stock_name.Close.shift(1)  #next rows value
    stock_name["price_shocks"] = (abs((stock_name["price_T+1"] - stock_name["Close"]) / stock_name["Close"] * 100)  > 2).astype(int)
    stock_name["black_swan"] = stock_name['price_shocks'] 
    
    return stock_name
```

```python
price_shock(tcs)
```

```python
price_shock(infy)
```

```python
price_shock(nifty)
```
